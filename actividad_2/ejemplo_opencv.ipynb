```python
#Importar librerias
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
bgr = cv2.imread('imagenes/imagen_1_actividad_2.jpeg') #Leer imagen

print("El tamaño de la imagen es:", bgr.shape) #Mostrar tamaño de la imagen

#convertir a RGB
rgb = cv2.cvtColor(bgr, cv2.COLOR_BGR2RGB)

plt.imshow(rgb) #Mostrar imagen
```

```python
#Convertir a escala de grises
gray = cv2.cvtColor(bgr, cv2.COLOR_BGR2GRAY)
```

```python
#Mostrar las componentes de color de la imagen
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(bgr[:,:,0], cmap='gray')
plt.title('Componente Azul')
plt.subplot(1,3,2)
plt.imshow(bgr[:,:,1], cmap='gray')
plt.title('Componente Verde')
plt.subplot(1,3,3)
plt.imshow(bgr[:,:,2], cmap='gray')
plt.title('Componente Roja')
plt.show()
```

```python
#Mostrar componentes de HLS de la imagen
hls = cv2.cvtColor(bgr, cv2.COLOR_BGR2HLS)
plt.figure(figsize=(10,5))
plt.subplot(1,3,1)
plt.imshow(hls[:,:,0], cmap='gray')
plt.title('Componente H')
plt.subplot(1,3,2)
plt.imshow(hls[:,:,1], cmap='gray')
plt.title('Componente L')
plt.subplot(1,3,3)
plt.imshow(hls[:,:,2], cmap='gray')
plt.title('Componente S')
plt.show()
```

```python
#Ajuste de brillo y contraste
alpha = 1.5 # Contraste
beta = 50   # Brillo
adjusted = cv2.convertScaleAbs(bgr, alpha=alpha, beta=beta)
plt.imshow(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB))
plt.title('Imagen con brillo y contraste ajustados')
```

```python
#Aplicar filtro Gaussiano
blurred = cv2.GaussianBlur(bgr, (15, 15), 0)
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
plt.title('Imagen con filtro Gaussiano')
plt.show()
```
