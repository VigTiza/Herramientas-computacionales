```python
# Transformaciones geométricas para cada imagen
imagenes = [bgr, bgr2, bgr3, bgr4, bgr5]
nombres = ['Imagen 1', 'Imagen 2', 'Imagen 3', 'Imagen 4', 'Imagen 5']

for idx, (img, nombre) in enumerate(zip(imagenes, nombres)):
    # Redimensionar (más grande)
    bigger = cv2.resize(img, (img.shape[1]*2, img.shape[0]*2), interpolation=cv2.INTER_LINEAR)
    # Redimensionar (más pequeña)
    smaller = cv2.resize(img, (max(1,img.shape[1]//2), max(1,img.shape[0]//2)), interpolation=cv2.INTER_AREA)
    # Rotar 45 grados y trasladar
    rows, cols = img.shape[:2]
    M_rot = cv2.getRotationMatrix2D((cols/2, rows/2), 45, 1)
    rotated = cv2.warpAffine(img, M_rot, (cols, rows))
    M_trans = np.float32([[1, 0, 50], [0, 1, 30]])
    translated = cv2.warpAffine(rotated, M_trans, (cols, rows))
    # Volteo horizontal y vertical
    flipped = cv2.flip(img, -1)  # -1: ambos ejes

    # Mostrar cada transformación en una figura separada para que se aprecie el cambio de tamaño
    plt.figure(figsize=(6,6))
    plt.title(f'{nombre} - Original')
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    plt.figure(figsize=(10,10))
    plt.title(f'{nombre} - Más grande')
    plt.imshow(cv2.cvtColor(bigger, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    plt.figure(figsize=(3,3))
    plt.title(f'{nombre} - Más pequeña')
    plt.imshow(cv2.cvtColor(smaller, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    plt.figure(figsize=(6,6))
    plt.title(f'{nombre} - Rotada+Trasladada')
    plt.imshow(cv2.cvtColor(translated, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    plt.figure(figsize=(6,6))
    plt.title(f'{nombre} - Volteo H/V')
    plt.imshow(cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```

```python
# Detección de bordes: Canny, Sobel y Laplaciano para cada imagen
imagenes = [bgr, bgr2, bgr3, bgr4, bgr5]
nombres = ['Imagen 1', 'Imagen 2', 'Imagen 3', 'Imagen 4', 'Imagen 5']

for idx, (img, nombre) in enumerate(zip(imagenes, nombres)):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    # Canny
    canny = cv2.Canny(gray, 100, 200)
    # Sobel
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)
    sobel = cv2.magnitude(sobelx, sobely)
    sobel = np.uint8(np.clip(sobel, 0, 255))
    # Laplaciano
    laplacian = cv2.Laplacian(gray, cv2.CV_64F)
    laplacian = np.uint8(np.clip(np.absolute(laplacian), 0, 255))

    plt.figure(figsize=(16,4))
    plt.suptitle(f'Detección de bordes - {nombre}', fontsize=16)
    plt.subplot(1,4,1)
    plt.title('Original (gris)')
    plt.imshow(gray, cmap='gray')
    plt.axis('off')
    plt.subplot(1,4,2)
    plt.title('Canny')
    plt.imshow(canny, cmap='gray')
    plt.axis('off')
    plt.subplot(1,4,3)
    plt.title('Sobel')
    plt.imshow(sobel, cmap='gray')
    plt.axis('off')
    plt.subplot(1,4,4)
    plt.title('Laplaciano')
    plt.imshow(laplacian, cmap='gray')
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
# Detección de esquinas (Harris y Shi-Tomasi) para todas las imágenes
imagenes = [bgr, bgr2, bgr3, bgr4, bgr5]
nombres = ['Imagen 1', 'Imagen 2', 'Imagen 3', 'Imagen 4', 'Imagen 5']

for img, nombre in zip(imagenes, nombres):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    gray_f = np.float32(gray)
    # Harris
    harris = cv2.cornerHarris(gray_f, 2, 3, 0.04)
    harris_img = img.copy()
    harris_img[harris > 0.01 * harris.max()] = [0, 0, 255]
    # Shi-Tomasi
    shi_corners = cv2.goodFeaturesToTrack(gray, 100, 0.01, 10)
    shi_img = img.copy()
    if shi_corners is not None:
        shi_corners = np.intp(shi_corners)
        for i in shi_corners:
            x, y = i.ravel()
            cv2.circle(shi_img, (x, y), 4, (0, 255, 0), -1)
    plt.figure(figsize=(12, 5))
    plt.suptitle(f'{nombre} - Detección de esquinas', fontsize=16)
    plt.subplot(1, 3, 1)
    plt.title('Original')
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.subplot(1, 3, 2)
    plt.title('Harris')
    plt.imshow(cv2.cvtColor(harris_img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.subplot(1, 3, 3)
    plt.title('Shi-Tomasi')
    plt.imshow(cv2.cvtColor(shi_img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
# Detección de formas: Círculos (HoughCircles) y contornos principales para todas las imágenes (optimizado)
imagenes = [bgr, bgr2, bgr3, bgr4, bgr5]
nombres = ['Imagen 1', 'Imagen 2', 'Imagen 3', 'Imagen 4', 'Imagen 5']

for img, nombre in zip(imagenes, nombres):
    # Redimensionar si la imagen es muy grande
    max_dim = 600
    scale = min(max_dim / img.shape[0], max_dim / img.shape[1], 1.0)
    if scale < 1.0:
        img_proc = cv2.resize(img, (int(img.shape[1]*scale), int(img.shape[0]*scale)))
    else:
        img_proc = img.copy()
    output_circles = img_proc.copy()
    output_contours = img_proc.copy()
    gray = cv2.cvtColor(img_proc, cv2.COLOR_BGR2GRAY)
    gray_blur = cv2.medianBlur(gray, 5)
    # Detección de círculos (ajustar parámetros para mayor velocidad)
    circles = cv2.HoughCircles(gray_blur, cv2.HOUGH_GRADIENT, dp=1.2, minDist=50, param1=60, param2=40, minRadius=20, maxRadius=0)
    if circles is not None:
        circles = np.uint16(np.around(circles))
        for i in circles[0, :]:
            cv2.circle(output_circles, (i[0], i[1]), i[2], (0, 255, 0), 2)
            cv2.circle(output_circles, (i[0], i[1]), 2, (0, 0, 255), 3)
    # Detección de contornos principales (solo los 10 más grandes)
    ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contours = sorted(contours, key=cv2.contourArea, reverse=True)[:10]
    cv2.drawContours(output_contours, contours, -1, (255, 0, 0), 2)
    # Mostrar resultados
    plt.figure(figsize=(15, 5))
    plt.suptitle(f'{nombre} - Detección de formas', fontsize=16)
    plt.subplot(1, 3, 1)
    plt.title('Original')
    plt.imshow(cv2.cvtColor(img_proc, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.subplot(1, 3, 2)
    plt.title('Círculos (Hough)')
    plt.imshow(cv2.cvtColor(output_circles, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.subplot(1, 3, 3)
    plt.title('Contornos principales')
    plt.imshow(cv2.cvtColor(output_contours, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```

```python
# 9. Segmentación de objetos: Umbralado (global y adaptativo), Segmentación por color (HSV), K-means clustering para todas las imágenes
imagenes = [bgr, bgr2, bgr3, bgr4, bgr5]
nombres = ['Imagen 1', 'Imagen 2', 'Imagen 3', 'Imagen 4', 'Imagen 5']

for idx, (img, nombre) in enumerate(zip(imagenes, nombres)):
    # Redimensionar para visualización rápida si es muy grande
    max_dim = 400
    scale = min(max_dim / img.shape[0], max_dim / img.shape[1], 1.0)
    if scale < 1.0:
        img_proc = cv2.resize(img, (int(img.shape[1]*scale), int(img.shape[0]*scale)))
    else:
        img_proc = img.copy()
    gray = cv2.cvtColor(img_proc, cv2.COLOR_BGR2GRAY)
    hsv = cv2.cvtColor(img_proc, cv2.COLOR_BGR2HSV)

    # --- Umbralado global ---
    _, th_global = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

    # --- Umbralado adaptativo ---
    th_adapt = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 11, 2)

    # --- Segmentación por color (HSV) ---
    # Rango ejemplo: colores rojos (ajustar según imagen)
    lower = np.array([0, 50, 50])
    upper = np.array([10, 255, 255])
    mask1 = cv2.inRange(hsv, lower, upper)
    lower2 = np.array([170, 50, 50])
    upper2 = np.array([180, 255, 255])
    mask2 = cv2.inRange(hsv, lower2, upper2)
    mask_hsv = cv2.bitwise_or(mask1, mask2)

    # --- K-means clustering ---
    Z = img_proc.reshape((-1,3))
    Z = np.float32(Z)
    K = 3
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 10, 1.0)
    ret,label,center=cv2.kmeans(Z,K,None,criteria,10,cv2.KMEANS_RANDOM_CENTERS)
    center = np.uint8(center)
    res = center[label.flatten()]
    kmeans_img = res.reshape((img_proc.shape))

    # --- Visualización comparativa ---
    plt.figure(figsize=(16,8))
    plt.suptitle(f'Segmentación - {nombre}', fontsize=16)
    plt.subplot(2,3,1)
    plt.title('Original')
    plt.imshow(cv2.cvtColor(img_proc, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.subplot(2,3,2)
    plt.title('Umbralado global')
    plt.imshow(th_global, cmap='gray')
    plt.axis('off')
    plt.subplot(2,3,3)
    plt.title('Umbralado adaptativo')
    plt.imshow(th_adapt, cmap='gray')
    plt.axis('off')
    plt.subplot(2,3,4)
    plt.title('Máscara HSV (rojo)')
    plt.imshow(mask_hsv, cmap='gray')
    plt.axis('off')
    plt.subplot(2,3,5)
    plt.title('K-means (K=3)')
    plt.imshow(cv2.cvtColor(kmeans_img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.tight_layout()
    plt.show()
```
